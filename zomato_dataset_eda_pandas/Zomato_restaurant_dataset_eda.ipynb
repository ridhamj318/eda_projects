```python
# How many restaurants are listed in each country?

final_df[['Restaurant ID','Country']].groupby(['Country']).count().sort_values(by='Restaurant ID', ascending=False)
```

```python
restaurant_count = final_df[['Restaurant ID','Country']].groupby(['Country']).count().sort_values(by='Restaurant ID', ascending=False).head(10)
restaurant_count
```

```python
# top 10 cities with highest restaurants
sns.barplot(data=restaurant_count,x = 'Country',y = 'Restaurant ID')
plt.title('Top 10 Cities with most Restaurants')
plt.show()
```

```python
# How many restaurants offer table booking vs. those that don’t?

table_booking = final_df['Has Table booking'].value_counts().reset_index()
table_booking
```

```python
# Do restaurants with table booking have higher average ratings?
avg_rating = final_df.groupby('Has Table booking')['Aggregate rating'].mean().reset_index()
avg_rating
```

```python
sns.barplot(data=avg_rating,x='Has Table booking', y= 'Aggregate rating')
plt.title('Average Rating by Table Booking Availability')
plt.show()
```

```python
# Do restaurants with table booking have higher average ratings?

votes = final_df.groupby('Has Online delivery')[['Votes','Aggregate rating']].mean().reset_index()
votes
```

```python
price_range = final_df.groupby('Price range')['Aggregate rating'].mean().reset_index()
price_range
```

```python
# Which cities have the most expensive and highly-rated restaurants?
high_rated = final_df[final_df['Aggregate rating'] >= 4.5]
avg_price = final_df['Average Cost for two'].mean()
expensive_restaurants = final_df[final_df['Average Cost for two'] > avg_price]
high_rated_restaurants = final_df[(final_df['Aggregate rating'] >= 4.5) &
    (final_df['Average Cost for two'] > avg_price)]

result = high_rated_restaurants['City'].value_counts().reset_index()
result.head(10)
```

```python
sns.barplot(data=result.head(10), x='City',y='count')
plt.title('Top 10 City with Expensive and High Rated Resaurant')
plt.show()
```
